```python
from gtts import gTTS

text_to_say = "A black dog and a tri-colored dog playing with each other on the road"

language = "en"

gtts_object = gTTS(text = text_to_say, lang = language, slow = False)

gtts_object.save("/content/gtts.wav")
```
